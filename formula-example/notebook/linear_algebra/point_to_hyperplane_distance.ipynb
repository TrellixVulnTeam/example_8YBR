```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
x, y = np.meshgrid(range(10), range(10))
z = np.zeros((10, 10))
ax.plot_surface(X = x, Y = y, Z = z, alpha=0.2)
ax.scatter(xs=8, ys=8, zs=8, c='red')
ax.text(x=8, y=8, z=8 + 0.5, s="B")
ax.scatter(xs=1, ys=1, zs=0, c='green')
ax.text(x=1, y=1, z= 0 - 1.5, s="A")
ax.quiver(1, 1, 0, 7, 7, 8, arrow_length_ratio = 0.1, color="blue", linestyle='dashed')
ax.quiver(1, 1, 0, 0, 0, 10, arrow_length_ratio = 0.1, color="green", linestyle='dashed')
ax.text(x=0.5, y=0, z=10, s="n")
ax.quiver(1, 1, 8, 7, 7, 0, arrow_length_ratio = 0, color="black", linestyle='dashed')
ax.text(x=-2, y=0, z=4, s="$proj_{n}(\\overrightarrow{AB})$")
ax.text(x=7, y=3, z=0, s="$\\vec n \\cdot \\vec x + b = 0$")
ax.set_xlim(-1,10)
ax.set_ylim(-1,10)
ax.set_zlim(-1,10)
plt.show()
```

Distance from point B to a hyperplane $n \cdot x + b = 0$ is:
$$\parallel proj_{n}(\overrightarrow{AB}) \parallel = \dfrac{\mid \vec n \cdot \overrightarrow{OB} + b \mid}{\parallel \vec n \parallel}$$

# proof

$\parallel proj_{n}(\overrightarrow{AB}) \parallel = \parallel \dfrac{\overrightarrow{AB} \cdot \vec n}{\vec n \cdot \vec n} \cdot \vec n \parallel$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<a href="https://github.com/hsiehpinghan/example/blob/master/formula-example/notebook/linear_algebra/projection_onto_a_vector.ipynb">projection onto a vector</a> <br><br> 
$= \dfrac{\mid \overrightarrow{AB} \cdot \vec n \mid}{\parallel \vec n \parallel^{2}} \cdot \parallel \vec n \parallel$ <br><br>
$= \dfrac{\mid \big( \overrightarrow{OB} - \overrightarrow{OA} \big) \cdot \vec n \mid}{\parallel \vec n \parallel}$ <br><br>
$\because \vec n \cdot \vec x + b = 0$ <br><br>
$\therefore \overrightarrow{OA} \cdot \vec n = -b$ <br><br>
$= \dfrac{\mid \vec n \cdot \overrightarrow{OB} + b \mid}{\parallel \vec n \parallel}$
